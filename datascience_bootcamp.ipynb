```python
units = nutrition.astype(str).replace('[^a-zA-Z]', '', regex=True).mode()
units
```

```python
columns = list(units.columns)
```

```python
column_units = list(units.iloc[0])
```

```python
nutrition.columns = [c + "_" + cu if cu != '' else c for c, cu in zip(columns, column_units)]
```
